```python
###load in the bike station data 
divvyStations =  pd.read_csv("/Users/irena/Psych Bikes/Downloads/Divvy_Trips_2017_Q3Q4/Divvy_Stations_2017_Q3Q4.csv")
divvyStations.shape
```

```python
#3 create some variables that denote weekday vs. weekend, hour of day (0-24) 

totalTrips['is_weekend'] =totalTrips['start_time'].dt.dayofweek > 4
totalTrips['hour'] = totalTrips['start_time'].dt.hour

#also create variables that will show if time of day is in the "morning rush" times (5-10 am) or "evening rush" (3-8 pm)
totalTrips['is_morning'] = (totalTrips['hour'] > 5)& (totalTrips['hour'] <10)
totalTrips['is_evening'] = (totalTrips['hour'] > 15)& (totalTrips['hour'] <20)
```

```python
## get the data ready to plot ONLY weekdays and subscriber (since the non-subscriber curves were not very interesting)
weekdayTrips= totalTrips[(totalTrips['is_weekend'] == False)&(totalTrips['usertype'] == 'Subscriber')]
```

```python
##function that takes in two datasets (a trip dataset and a station location dataset)
## the function calculates total arrivals & total departures and joins the two datasets together 
## main source code from : https://medium.com/@drevets/mapping-divvy-data-using-python-156e2a4dc59d - credit to Emily Drevets
def add_trip_counts_to_stations(stations, trips):
    '''
    input: data frames
    output: data frame
    '''
    departure_counts = trips.groupby('from_station_id').count()
    departure_counts = departure_counts.iloc[:, [0]]
    departure_counts.columns = ['departure_count']

    arrival_counts = trips.groupby('to_station_id').count()
    arrival_counts = arrival_counts.iloc[:, [0]]
    arrival_counts.columns = ['arrival_count']

    stations = pd.merge(departure_counts, stations,
                        right_on='id',
                        left_index=True).merge(arrival_counts,
                                               left_on='id',
                                               right_index=True)
    stations['total_activity'] = stations['arrival_count'] + stations['departure_count'] ##total activity 
    stations['net_activity'] = stations['arrival_count'] - stations['departure_count'] ## are people arriving or departing more 
    stations['arrive_depart_ratio'] = stations['arrival_count']/stations['departure_count'] ## total activity as a % of 
    ## returns a dataset that has these three new variables 
    return stations
```

```python
## morning trips dataset 
morningWeekday = add_trip_counts_to_stations(divvyStations, weekdayTrips[weekdayTrips['is_morning']==True])
## evening trips dataset
eveningWeekday = add_trip_counts_to_stations(divvyStations, weekdayTrips[weekdayTrips['is_evening']==True])
```

```python
##prepare morning and evening data in order to create a scatterplot: 
morningData = morningWeekday[['id', 'name','arrive_depart_ratio']]
eveningData = eveningWeekday[['id', 'name','arrive_depart_ratio']]
```

```python
##rename the ratio columns and then merge the datasets: 
morningData =morningData.rename(columns = {'arrive_depart_ratio':'morning_arrive_depart_ratio'})
eveningData =eveningData.rename(columns = {'arrive_depart_ratio':'evening_arrive_depart_ratio'})

graphData = pd.merge(morningData, eveningData, on=['id', 'name'], how='inner')
```

```python
###create scatterplot of the ratios: 
plt.scatter(graphData.morning_arrive_depart_ratio,graphData.evening_arrive_depart_ratio)
```

```python
##function that takes in a dataset and returns a folium map showing arrival and departure count, and activity percentage
## I got this from: https://medium.com/@drevets/mapping-divvy-data-using-python-156e2a4dc59d - credit to Emily Drevets
def put_stations_on_map(stations):
    '''
    input: data frame
    output: folium map
    '''
    map = folium.Map(location=[41.88, -87.62],
                            zoom_start=13,
                            tiles="CartoDB dark_matter")
    for index, station in stations.iterrows():
        popup_text = "{}<br> Total departures: {}<br> Total arrivals: {}<br> Total Activity % : {}<br>"
        popup_text = popup_text.format(stations.at[index,    "name"], stations.at[index, "departure_count"], stations.at[index, "arrival_count"],
                                      round(100*stations.at[index, "activity_percentage"], 2))
        folium.CircleMarker(location=[stations.at[index, 'latitude'], stations.at[index, 'longitude']],
                            radius=2, color="white",popup=popup_text).add_to(map)
    return map
```

```python
##function that takes in a dataset and returns a folium map showing arrival and departure count, and net activity
## I got this from: https://medium.com/@drevets/mapping-divvy-data-using-python-156e2a4dc59d - credit to Emily Drevets
def map_station_net_activity(stations):
    '''
    input: data frame
    output: folium map
    '''
    map = folium.Map(location=[41.88, -87.62],
                            zoom_start=13,
                            tiles="CartoDB dark_matter")
    for index, station in stations.iterrows():
        popup_text = "{}<br> Total arrivals: {} <br> Total departures: {}<br> Net Activity : {}<br>"
        popup_text = popup_text.format(stations.at[index,    "name"], stations.at[index, "arrival_count"],
                                       stations.at[index, "departure_count"],
                                       stations.at[index, "net_activity"])
        folium.CircleMarker(location=[stations.at[index, 'latitude'], stations.at[index, 'longitude']],
                            radius=2, color="white",popup=popup_text).add_to(map)
    return map
```

```python
##map that shows total departures, total arrivals, and activity percentage for morning trips: 
morningMap = put_stations_on_map(morningWeekday)
morningMap.add_child(plugins.HeatMap(morningWeekday[['latitude', 'longitude', 'activity_percentage']].as_matrix(), 
                                     radius = 16))
```

```python
## find % of stations that have negative net activity in the morning
len(morningWeekday.loc[morningWeekday.net_activity < 0])/584
## over 50% of bike stations have more people departing than arriving
## this could indicate that subscribers are going to select stations in the morning
```

```python
##map that shows total departures, total arrivals, and net activity
morningNetActivity = map_station_net_activity(morningWeekday)
morningNetActivity.add_child(plugins.HeatMap(morningWeekday[['latitude', 'longitude', 'net_activity']].as_matrix(), 
                                     radius = 16, gradient = {.590: 'blue', 1: 'yellow'}))
```

```python
def add_trip_counts_to_stations(stations, trips):
    '''
    input: data frames
    output: data frame
    '''
    departure_counts = trips.groupby(['from_station_id', 'week_num']).count()
    departure_counts = departure_counts.iloc[:, [0]]
    departure_counts.columns = ['departure_count']
    

    arrival_counts = trips.groupby(['to_station_id', 'week_num']).count()
    arrival_counts = arrival_counts.iloc[:, [0]]
    arrival_counts.columns = ['arrival_count']
    

    stations = pd.merge(departure_counts, stations,left_on='from_station_id',
                        right_on='id').merge(arrival_counts,
                                               left_on='id', right_on='from_station_id', right_index=True)
    stations['total_activity'] = stations['arrival_count'] + stations['departure_count'] ##total activity 
    stations['departure_arrival_ratio'] = stations['departure_count']/stations['arrival_count'] ## are people arriving or departing more 
    stations['activity_percentage'] = stations['total_activity']/stations['total_activity'].sum()
    
    return stations
```

```python
def put_stations_on_map(stations):
    '''
    input: data frame
    output: folium map
    '''
    map = folium.Map(location=[41.88, -87.62],
                            zoom_start=13,
                            tiles="CartoDB dark_matter")
    for index, station in stations.iterrows():
        popup_text = "{}<br> Total departures: {}<br> Total arrivals: {}<br> Total Activity % : {}<br>"
        popup_text = popup_text.format(stations.at[index,    "name"], stations.at[index, "departure_count"], stations.at[index, "arrival_count"],
                                      round(100*stations.at[index, "activity_percentage"], 2))
        folium.CircleMarker(location=[stations.at[index, 'latitude'], stations.at[index, 'longitude']],
                            radius=2, color="white",popup=popup_text).add_to(map)
    return map
```

```python
def map_station_net_activity(stations):
    '''
    input: data frame
    output: folium map
    '''
    map = folium.Map(location=[41.88, -87.62],
                            zoom_start=13,
                            tiles="CartoDB dark_matter")
    for index, station in stations.iterrows():
        popup_text = "{}<br> Total arrivals: {} <br> Total departures: {}<br> Net Activity : {}<br>"
        popup_text = popup_text.format(stations.at[index,    "name"], stations.at[index, "arrival_count"],
                                       stations.at[index, "departure_count"],
                                       stations.at[index, "net_activity"])
        folium.CircleMarker(location=[stations.at[index, 'latitude'], stations.at[index, 'longitude']],
                            radius=2, color="white",popup=popup_text).add_to(map)
    return map
```

```python
##calculate distance between the to and from stations using the lat/long coords: 

def dist_from_coordinates(lat1, lon1, lat2, lon2):
  R = 6371  # Earth radius in km

  #conversion to radians
  d_lat = np.radians(lat2-lat1)
  d_lon = np.radians(lon2-lon1)

  r_lat1 = np.radians(lat1)
  r_lat2 = np.radians(lat2)

  #haversine formula
  a = np.sin(d_lat/2.) **2 + np.cos(r_lat1) * np.cos(r_lat2) * np.sin(d_lon/2.)**2

  haversine = 2 * R * np.arcsin(np.sqrt(a))

  return haversine
```
